```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('ShaunTest').getOrCreate()
```

```python
from pyspark import SparkContext, SparkConf

spark_conf = SparkConf()\
  .setAppName('ShaunTest')\
  .setMaster("local[*]")

sc = SparkContext.getOrCreate(spark_conf)
```

```python
sc
```

```python
nums = list(range(0, 1000001))
len(nums)
```

```python
nums_rdd = sc.parallelize(nums)
nums_rdd
```

```python
nums_rdd.take(5)
```

```python
squared_nums_rdd = nums_rdd.map(lambda x: x ** 2)
squared_nums_rdd.take(5)
```

```python
pairs = squared_nums_rdd.map(lambda x: (x,len(str(x))))
pairs.take(25)
```

```python
pairsEvenDigit = pairs.filter(lambda x: (x[1] % 2) == 0 )
pairsEvenDigit.take(25)
```

```python
flipped = pairsEvenDigit.map(lambda x: (x[1],x[0]))
flipped.take(25)
```

```python
grouped = flipped.groupByKey()
grouped.take(25)
```

```python
grouped = grouped.map(lambda x: (x[0],list(x[1])))
grouped.take(4)
```

```python
Avggrouped = grouped.map(lambda x: (x[0],sum(x[1])/len(x[1])))
Avggrouped.take(5)
```

```python
Avggrouped.collect()
```

```python
betterway=flipped.reduceByKey(lambda x,y: (x+y))
betterway.take(25)
```
